# EDA (Exploratory Data Ananlysis)

1. Departure Time,
2. Arival Time,
3. Date of Journey,
4. Duration 

```python
train_data['Journey_day']=pd.to_datetime(train_data['Date_of_Journey'],format="%d/%m/%Y").dt.day
train_data['Journey_month']=pd.to_datetime(train_data['Date_of_Journey'],format="%d/%m/%Y").dt.month
train_data.drop(['Date_of_Journey'],axis=1,inplace=True)  # Drop column after conversion
```

```python
train_data.head()
```

```python
#Departure Time EDA
train_data['Dep_hour']=pd.to_datetime(train_data['Dep_Time']).dt.hour
train_data['Dep_min']=pd.to_datetime(train_data['Dep_Time']).dt.minute
train_data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#Arrival Time EDA
train_data['Arrival_hour']=pd.to_datetime(train_data['Arrival_Time']).dt.hour
train_data['Arrival_min']=pd.to_datetime(train_data['Arrival_Time']).dt.minute
train_data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#Duration of journey

duration=list(train_data['Duration'])

for i in range(len(duration)):
    if len(duration[i].split()) !=2:
        if "h" in duration[i]:
            duration[i]=duration[i].strip()+" 0m"
        else:
            duration[i]="0h "+duration[i]
        
duration_hours=[]
duration_min=[]
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep="h")[0]))
    duration_min.append(int(duration[i].split(sep="m")[0].split()[-1]))   
```

```python
train_data['Duration_hours']=duration_hours
train_data['Duration_mins']=duration_min
train_data.drop(['Duration'],axis=1,inplace=True)
```

```python
train_data.head()
```

# Handling Categorical Variables
1. Nominal data --> data are not in any order ---> OneHotEncoder in this case
2. Ordinal data --> data in order --> LableEncoder in this case

## Ordinal Data is where the data is ranked like grades
## nominal data were there is not superior to other data

```python
train_data['Airline'].value_counts()   ##Nominal Data--> OneHotEncoding
```

```python
sns.catplot(y="Price", x="Airline", data=train_data.sort_values("Price",ascending=False),kind='boxen',height=6,aspect=3)
plt.show()
```

```python
Airline =train_data[['Airline']]
Airline =pd.get_dummies(Airline,drop_first=True)
Airline.head()
```

```python
train_data['Source'].value_counts()
```

```python
sns.catplot(y="Price", x="Source", data=train_data.sort_values("Price",ascending=False),kind='boxen',height=6,aspect=3)
plt.show()
```

```python
Source =train_data[['Source']]
Source =pd.get_dummies(Source,drop_first=True)
Source.head()
```

```python
train_data['Destination'].value_counts()
```

```python
sns.catplot(y="Price", x="Destination", data=train_data.sort_values("Price",ascending=False),kind='boxen',height=6,aspect=3)
plt.show()
```

```python
Destination =train_data[['Destination']]
Destination =pd.get_dummies(Destination,drop_first=True)
Destination.head()
```

```python
train_data['Route']
```

```python
# Additional_Info contains lmost 80% no_info
# Route and Total_Stops are related to each other

train_data.drop(['Route',"Additional_Info"],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
train_data['Total_Stops'].value_counts()
```

```python
train_data.replace({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
data_train=pd.concat([train_data,Airline,Source,Destination],axis=1)
```

```python
data_train.head()
```

```python
data_train.drop(['Airline','Source','Destination'],axis=1,inplace=True)
```

```python
data_train.head()
```
